```python
tempp=[32.5,31.8,33.0,35.2,36.6]
total=0
for temp in tempp:
    total=total+temp
    average=total/len(tempp)
print(average)
```

```python
import numpy as np
tempp=np.array([32.5,31.8,33.0,35.2,36.6] )
average=np.mean(tempp)
print(average)
```
