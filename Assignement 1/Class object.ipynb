```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Amit", 20)
print(s1.name, s1.age)
```

```python
class College:
    name = "ABC College"

    @classmethod
    def show_name(cls):
        print(cls.name)

College.show_name()
```

```python
class College:
    name = "ABC College"

    @classmethod
    def show_name(cls):
        print(cls.name)

College.show_name()
```

```python
class Math:
    @staticmethod
    def add(a, b):
        print(a + b)

Math.add(5, 10)
```

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name

s = Student("Vivek")
print(s)
```

```python
class Test:
    def __del__(self):
        print("Object destroyed")

obj = Test()
del obj
```

```python
class Parent:
    def show(self):
        print("This is parent class")

class Child(Parent):
    pass

obj = Child()
obj.show()
```

```python
class Parent:
    def show(self):
        print("Parent method")

class Child(Parent):
    def show(self):
        print("Child method")

obj = Child()
obj.show()
```
